### Data Import and Cleaning

We start by importing the Titanic dataset from the `data/raw/` folder.  
Before modeling, we must handle missing values to avoid errors or biased results.  

Steps taken:  
- **Checked missing values** in each column.  
- **Created a copy** of the DataFrame (`df_clean`) to preserve the raw data (`df`).  
- **Handled missing values**:  
  - Filled `Age` with the median (robust to outliers).  
  - Dropped the `Cabin` column (too many missing values).  
  - Filled `Embarked` with the mode (most common port).  

After this step, the dataset is clean and ready for feature engineering.  


```python
##importing data

import pandas as pd 

df = pd.read_csv("../data/raw/titanic.csv")

#------

#checking missing values in each column

df.isnull().sum()

#------

## making a copy of DF // we can compare later raw vs cleaned
df_clean = df.copy()

## filing missing age with median
df_clean["Age"] = df_clean["Age"].fillna(df_clean["Age"].median())

## Drop cabin colummn
df_clean.drop(columns = ["Cabin"], inplace = True)

## filling missing embarked with mode
df_clean["Embarked"] = df_clean["Embarked"].fillna(df_clean["Embarked"].mode()[0])

df_clean.isnull().sum()


```
