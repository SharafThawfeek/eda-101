### Load saved metrics

We read `reports/metrics.json` written in the modeling notebook.  



```python
import os, json
import pandas as pd

metrics_path = "../reports/metrics.json"

metrics = None
if os.path.exists(metrics_path):
    with open(metrics_path, "r") as f:
        metrics = json.load(f)
    
    preferred_order = ["accuracy", "precision", "recall", "f1"]
    cols = [k for k in preferred_order if k in metrics]
    metrics_df = pd.DataFrame([[metrics.get(k) for k in cols]], columns=cols).round(3)
    display(metrics_df)
else:
    print("metrics.json not found at", metrics_path, "\nRun 03_modeling.ipynb and save metrics first.")
```

### Confusion Matrix

We display the confusion matrix image saved by `03_modeling.ipynb`.  
Darker squares on the diagonal = more correct predictions.



```python
from IPython.display import Image, display

cm_path = "../reports/figures/confusion_matrix.png"

if os.path.exists(cm_path):
    display(Image(filename=cm_path))
else:
    print("Missing:", cm_path, "- run the confusion matrix cell in 03_modeling.ipynb")
```

### Key Takeaways (human summary)

- **Where it succeeds:** Correctly identifies many non-survivors/survivors (strong diagonal in confusion matrix).
- **Where it struggles:** Some false positives/negatives remain — normal for a simple baseline.
